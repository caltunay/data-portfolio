# Get packages & variables

```python
import pandas as pd
import numpy as np
from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns

from lifelines import KaplanMeierFitter
from lifelines.statistics import logrank_test


all_freqs_path = 'all_frequencies_days.csv'
mau_path = 'mau.csv'
hpath = 'histogram.csv'
lness_path = 'lness_over_time.csv'
cohort_path = 'cohort_activity.csv'
users_path = 'users_dim.csv'

columns_for_survival = ['friends_enabled', 'profile_photo_uploaded', 'android_pay_activated', 'offered_overdraft', 'active_first_week', 'age_group']
```

# Plot ECDF  

## Plot All Frequencies ECDF

```python
def load_and_preprocess_freqs(all_freqs_path):
    """
    Load and preprocess the frequency data.
    
    Parameters:
    all_freqs_path (str): The path to the frequency data CSV file.
    
    Returns:
    pd.DataFrame: The preprocessed frequency data with ECDF values.
    """
    freqs = pd.read_csv(all_freqs_path)
    
    # Sort the data by frequency_in_day
    freqs_sorted = freqs.sort_values(by='frequency_in_day').reset_index(drop=True)
    
    # Calculate ECDF values
    freqs_sorted['ecdf'] = np.arange(1, len(freqs_sorted) + 1) / len(freqs_sorted)
    
    return freqs_sorted

def plot_ecdf(freqs_sorted):
    """
    Plot the ECDF with annotations for the 90th percentile.
    
    Parameters:
    freqs_sorted (pd.DataFrame): The preprocessed frequency data with ECDF values.
    """
    # Plot the ECDF
    plt.figure(figsize=(7, 7))
    plt.step(freqs_sorted['frequency_in_day'], freqs_sorted['ecdf'], where='post', color='blue', label='ECDF')
    
    # Calculate the 90th percentile cut-off point
    cut_off_days_frequency = round(freqs_sorted[freqs_sorted['ecdf'] > .9]['frequency_in_day'].values[0])
    
    # Add a vertical line at the 90th percentile
    plt.axhline(y=.9, color='red', linestyle='--', label='90th Percentile')
    
    # Add an annotation for the 90th percentile line
    plt.annotate(
        f'{cut_off_days_frequency} days',
        xy=(cut_off_days_frequency, 0.90),  # Coordinates for annotation
        xytext=(cut_off_days_frequency + 10, 0.8),  # Text position
        arrowprops=dict(arrowstyle='->', facecolor='black')
    )
    
    # Adding labels and title
    plt.xlabel('# Days Since Last Active', fontsize=12)
    plt.ylabel('ECDF', fontsize=12)
    plt.title('ECDF for # Days Since Last Active', fontsize=15)
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # Set x-ticks
    x_ticks_ecdf = list(range(0, 400, 50))
    x_ticks_ecdf[0] = 1
    plt.xticks(x_ticks_ecdf, fontsize=12)
    plt.yticks(fontsize=12)
    plt.legend()
    
    # Show the plot
    plt.show()

# Main function to execute the workflow
def main(all_freqs_path):
    """
    Main function to execute the workflow.
    
    Parameters:
    all_freqs_path (str): The path to the frequency data CSV file.
    """
    freqs_sorted = load_and_preprocess_freqs(all_freqs_path)
    plot_ecdf(freqs_sorted)

# plot chart
main(all_freqs_path)
```

## # of Days Since Last Active Histogram

```python
def load_histogram_data(hpath):
    """
    Load the histogram data from a CSV file.
    
    Parameters:
    hpath (str): The path to the histogram data CSV file.
    
    Returns:
    pd.DataFrame: The loaded histogram data.
    """
    hist_df = pd.read_csv(hpath)
    return hist_df

def plot_histogram(hist_df):
    """
    Plot a histogram of the number of days since last active.
    
    Parameters:
    hist_df (pd.DataFrame): The histogram data.
    """
    plt.figure(figsize=(5, 5))
    plt.bar(hist_df['frequency'], hist_df['user_count'], color='skyblue')
    
    # Add title and labels
    plt.title('# Days Since Last Active Histogram', fontsize=16)
    plt.xlabel('# of Days Since Last Active', fontsize=12)
    plt.ylabel('User Count', fontsize=12)
    plt.xticks(list(range(0, 220, 20)), rotation=45, fontsize=12)  # Rotate x labels for better readability and increase font size
    
    # Show the plot
    plt.show()

# Main function to execute the workflow
def main(hpath):
    """
    Main function to execute the workflow.
    
    Parameters:
    hpath (str): The path to the histogram data CSV file.
    """
    hist_df = load_histogram_data(hpath)
    plot_histogram(hist_df)

#plot_chart'
main(hpath)
```

## Plot Avg Frequency ECDF

```python
def calculate_average_frequency(freqs):
    """
    Calculate the average frequency in days for each user.
    
    Parameters:
    freqs (pd.DataFrame): The frequency data.
    
    Returns:
    pd.DataFrame: The average frequency data for each user.
    """
    avg_freqs = freqs.groupby('user_id', as_index=False).agg({'frequency_in_day': np.mean})
    
    # Sort the data by average frequency_in_day
    avg_freqs_sorted = avg_freqs.sort_values(by='frequency_in_day').reset_index(drop=True)
    
    # Calculate ECDF values
    avg_freqs_sorted['ecdf'] = np.arange(1, len(avg_freqs_sorted) + 1) / len(avg_freqs_sorted)
    
    return avg_freqs_sorted

def plot_ecdf_avg_frequency(avg_freqs_sorted):
    """
    Plot the ECDF for the average frequency in days with annotations for the 90th percentile.
    
    Parameters:
    avg_freqs_sorted (pd.DataFrame): The sorted average frequency data with ECDF values.
    """
    # Plot the ECDF
    plt.figure(figsize=(7, 7))
    plt.step(avg_freqs_sorted['frequency_in_day'], avg_freqs_sorted['ecdf'], where='post', color='blue', label='ECDF')
    
    # Calculate the 90th percentile cut-off point
    cut_off_days_avg_frequency = round(avg_freqs_sorted[avg_freqs_sorted['ecdf'] > .9]['frequency_in_day'].values[0])
    
    # Add a horizontal line at the 90th percentile
    plt.axhline(y=.9, color='red', linestyle='--', label='90th Percentile')
    
    # Add an annotation for the 90th percentile line
    plt.annotate(
        f'{cut_off_days_avg_frequency} days',
        xy=(cut_off_days_avg_frequency, 0.90),  # Coordinates for annotation
        xytext=(cut_off_days_avg_frequency + 10, 0.8),  # Text position
        arrowprops=dict(arrowstyle='->', facecolor='black')
    )
    
    # Adding labels and title
    plt.xlabel('Days Since Last Active', fontsize=12)
    plt.ylabel('ECDF', fontsize=12)
    plt.title('ECDF for Average # Days Since Last Active per User', fontsize=15)
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # Set x-ticks
    x_ticks_ecdf_avg = list(range(0, 250, 50))
    x_ticks_ecdf_avg[0] = 1
    plt.xticks(x_ticks_ecdf_avg, fontsize=12)
    plt.yticks(fontsize=12)
    plt.legend()
    
    # Show the plot
    plt.show()

# Main function to execute the workflow
def main(freqs):
    """
    Main function to execute the workflow.
    
    Parameters:
    freqs (pd.DataFrame): The frequency data.
    """
    avg_freqs_sorted = calculate_average_frequency(freqs)
    plot_ecdf_avg_frequency(avg_freqs_sorted)

# read and plot usage
freqs = pd.read_csv(all_freqs_path)
main(freqs)
```

# L0 L1 & L2 Metrics 

## L0 Metric: MoM Change of MAU Ratio

```python
def load_and_preprocess_data(mau_path):
    """
    Load and preprocess the MAU data.
    
    Parameters:
    mau_path (str): The path to the MAU data CSV file.
    
    Returns:
    pd.DataFrame: The preprocessed MAU data.
    """
    mau_df = pd.read_csv(mau_path)
    
    # Convert date to correct format and filter for date when users hit 1000 to avoid volatility of early adopters
    mau_df['date'] = pd.to_datetime(mau_df['day'])
    mau_df_filtered = mau_df[mau_df['date'] > '2017-12-06'] 
    
    # Sort by date for visualization
    mau_df_filtered = mau_df_filtered.sort_values(by='date', ascending=True)
    mau_df_filtered.reset_index(drop=True, inplace=True)
    
    return mau_df_filtered

def calculate_mom_diff(mau_path):
    """
    Calculate the Month-over-Month (MoM) difference in MAU percentage.
    
    Parameters:
    mau_df_filtered (pd.DataFrame): The filtered MAU data.
    
    Returns:
    pd.DataFrame: The MAU data with MoM differences.
    """
    mau_df_filtered = load_and_preprocess_data(mau_path)
    mau_df_filtered.set_index('date', inplace=True)
    last_day_of_month = mau_df_filtered.resample('M').last()
    last_day_of_month['mom_diff'] = last_day_of_month['mau_pct'].diff()
    last_day_of_month['mom_diff'].fillna(0, inplace=True)
    last_day_of_month.reset_index(inplace=True)
    
    return last_day_of_month

def plot_mom_mau_ratio(last_day_of_month):
    """
    Plot the Month-over-Month (MoM) MAU Ratio change.
    
    Parameters:
    last_day_of_month (pd.DataFrame): The MAU data with MoM differences.
    """
    plt.figure(figsize=(12, 6))
    
    # Color bars based on positive/negative values
    colors = ['green' if val >= 0 else 'red' for val in last_day_of_month['mom_diff']]
    
    bars = plt.bar(last_day_of_month['date'], last_day_of_month['mom_diff'], color=colors, alpha=0.7, width=20)
    
    # Remove gridlines
    plt.grid(False)
    
    # Add a horizontal line at y=0
    plt.axhline(0, color='black', linewidth=1, linestyle='--')
    
    # Add labels with integer percentage difference
    for bar, value in zip(bars, last_day_of_month['mom_diff']):
        # Calculate label position based on the value
        label_y = bar.get_height() + 0.1 if value >= 0 else bar.get_height() - 0.1
        color_y = 'green' if value >= 0 else 'red'
        plt.text(
            bar.get_x() + bar.get_width() / 2,
            label_y,
            f'{int(value)}%',
            ha='center',
            va='bottom' if value >= 0 else 'top',
            color=color_y,
            fontsize=12,  # Increase font size
            fontweight='bold'  # Make font bold
        )
    
    plt.xlabel('Month')
    plt.ylabel('MoM MAU Ratio (%)')
    plt.title('Month-over-Month MAU Ratio Change')
    plt.xticks(ticks=last_day_of_month['date'], labels=[d.strftime('%Y-%m') for d in last_day_of_month['date']])
    plt.tight_layout()
    plt.show()

# Main function to execute the workflow
def main(mau_df_filtered):
    """
    Main function to execute the workflow.
    
    Parameters:
    mau_df_filtered (pd.DataFrame): The filtered MAU data.
    """
    last_day_of_month = calculate_mom_diff(mau_df_filtered)
    plot_mom_mau_ratio(last_day_of_month)

# plot chart
mau_path = 'mau.csv'
main(mau_path)
```

## L1 Metric: MAU Ratio by Day

```python
def plot_mau_ratio(mau_df_filtered):
    """
    Plot the MAU ratio over time with annotations.
    
    Parameters:
    mau_df_filtered (pd.DataFrame): The preprocessed MAU data.
    """
    # Create figure and axis objects
    fig, ax1 = plt.subplots(figsize=(14, 7))
    
    # Adding trendline for MAU percentage
    ax1.plot(mau_df_filtered['date'], mau_df_filtered['mau_pct'], color='blue', linestyle='-', label='MAU Ratio')
    
    # Setting up the y-axis
    ax1.set_ylabel('Percentage', color='black')
    ax1.tick_params(axis='y', labelcolor='black')
    ax1.legend(loc='upper right')
    ax1.grid(True, which='both', linestyle='--', alpha=.5)
    
    # Adding data labels and markers for the first day of each month
    first_day_of_month = mau_df_filtered[mau_df_filtered['date'].dt.is_month_start]
    
    for idx, row in first_day_of_month.iterrows():
        ax1.annotate(f"{row['mau_pct']:.0f}%", 
                     (row['date'], row['mau_pct']),
                     textcoords="offset points",
                     xytext=(0, 10),
                     ha='center')
        ax1.scatter(row['date'], row['mau_pct'], color='blue', s=20, zorder=5)  # Marker for the first day of the month
    
    # Adding data label and marker for the last available date
    last_date = mau_df_filtered['date'].iloc[-1]
    last_value = mau_df_filtered['mau_pct'].iloc[-1]
    
    ax1.annotate(f"{last_value:.2f}", 
                 (last_date, last_value),
                 textcoords="offset points",
                 xytext=(0, 10),
                 ha='center')
    ax1.scatter(last_date, last_value, color='blue', s=20, zorder=5)  # Marker for the last date
    
    # Title and layout adjustments
    plt.title('MAU Ratio')
    fig.tight_layout()
    
    # Show the plot
    plt.show()

# Main function to execute the workflow
def main(mau_path):
    """
    Main function to execute the workflow.
    
    Parameters:
    mau_path (str): The path to the MAU data CSV file.
    """
    mau_df_filtered = load_and_preprocess_data(mau_path)
    plot_mau_ratio(mau_df_filtered)

# plot the chart
main(mau_path)
```

## L2 Metric: Lness 2+/30 Ratio by Day

```python
def load_and_preprocess_lness_data(lness_path):
    """
    Load and preprocess the Lness data.
    
    Parameters:
    lness_path (str): The path to the Lness data CSV file.
    
    Returns:
    pd.DataFrame: The preprocessed Lness data.
    """
    lness_df = pd.read_csv(lness_path)
    lness_df['date'] = pd.to_datetime(lness_df['snapshot_date'])
    lness_filtered = lness_df[(lness_df['date'] > '2017-12-06') & (lness_df['date'] <= '2018-06-30')] # filtering out first 1000 users due to volatility
    lness_filtered.sort_values(by=['date'], ascending=True, inplace=True)
    return lness_filtered

def plot_lness_segment_ratios(lness_filtered):
    """
    Plot the Lness segment ratios over time with annotations for Mondays.
    
    Parameters:
    lness_filtered (pd.DataFrame): The preprocessed Lness data.
    """
    plt.figure(figsize=(14, 8))
    
    # Define colors for each segment
    segment_colors = {
        'Low Engagement': 'orange',
        'Healthy Engagement': 'green',
        'No Engagement': 'red'
    }
    
    # Plot each segment with the defined colors
    for segment in lness_filtered['lness_segment'].unique():
        segment_data = lness_filtered[lness_filtered['lness_segment'] == segment]
        plt.plot(segment_data['date'], segment_data['lness_segment_ratio'], label=segment, color=segment_colors[segment], marker='.')
        
        # Add data labels for Mondays with percentage values
        for i, row in segment_data.iterrows():
            if row['date'].dayofweek == 0:
                plt.text(row['date'], row['lness_segment_ratio'] + (1 if row['lness_segment_ratio'] >= 0 else -1.5), 
                         f"{int(row['lness_segment_ratio'])}%", color='black', ha='center', fontsize=9, fontweight='bold')
    
    # Add titles and labels
    plt.title('Lness Segment Ratios Over Time')
    plt.xlabel('Date')
    plt.ylabel('Lness Segment Ratio (%)')
    
    # Add a legend to differentiate segments
    plt.legend(title='Lness Segment')
    
    # Add grid lines for better readability
    plt.grid(True)
    
    # Set y-axis limits and ticks from 0 to 100
    plt.yticks(range(0, 101, 10))
    
    # Add a tight layout to ensure labels fit well
    plt.tight_layout()
    
    # Show the plot
    plt.show()

# Main function to execute the workflow
def main(lness_path):
    """
    Main function to execute the workflow.
    
    Parameters:
    lness_path (str): The path to the Lness data CSV file.
    """
    lness_filtered = load_and_preprocess_lness_data(lness_path)
    plot_lness_segment_ratios(lness_filtered)

# plot chart
main(lness_path)
```

# User Engagement Over Time

```python
def add_day_of_week_column(mau_df_filtered):
    """
    Add a column for the day of the week to the MAU DataFrame.
    
    Parameters:
    mau_df_filtered (pd.DataFrame): The filtered MAU data.
    
    Returns:
    pd.DataFrame: The MAU data with an additional day_of_week column.
    """
    
    mau_df_filtered['day_of_week'] = mau_df_filtered['date'].dt.day_name()
    return mau_df_filtered

def plot_mau_and_user_count(mau_df_filtered):
    """
    Plot the total user count and MAU percentage with a trendline.
    
    Parameters:
    mau_df_filtered (pd.DataFrame): The filtered MAU data.
    """
    # Create figure and axis objects
    fig, ax1 = plt.subplots(figsize=(14, 7))
    
    # Plotting total user count as bars
    ax1.bar(mau_df_filtered['date'], mau_df_filtered['total_users_at_date'], color='skyblue', alpha=0.6, label='Total User Count')
    
    # Setting up the first y-axis
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Total User Count', color='b')
    ax1.tick_params(axis='y', labelcolor='b')
    ax1.legend(loc='upper left')
    
    # Creating a second y-axis
    ax2 = ax1.twinx()
    
    # Plotting MAU percentage as a line
    ax2.plot(mau_df_filtered['date'], mau_df_filtered['mau_pct'], color='darkorange', label='MAU Percentage')
    
    # Adding trendline for MAU percentage
    ax2.plot(mau_df_filtered['date'], mau_df_filtered['rolling_30_day_mau_pct'], color='green', linestyle='-', label='MAU-30-Day-Trend')
    
    # Setting up the second y-axis
    ax2.set_ylabel('Percentage', color='black')
    ax2.tick_params(axis='y', labelcolor='black')
    ax2.legend(loc='upper right')
    
    ax2.grid(True, which='both', linestyle='--', alpha=.5)
    
    # Title and layout adjustments
    plt.title('Total Users, MAU % and MAU-30-Day-Trend by Day')
    fig.tight_layout()
    
    # Show the plot
    plt.show()

# Main function to execute the workflow
def main(mau_path):
    """
    Main function to execute the workflow.
    
    Parameters:
    mau_df_filtered (pd.DataFrame): The filtered MAU data.
    """
    mau_df_filtered = load_and_preprocess_data(mau_path)
    mau_df_filtered = add_day_of_week_column(mau_df_filtered)
    plot_mau_and_user_count(mau_df_filtered)

# Plot chart
main(mau_path)
```

# Cohorts 

```python
def load_and_preprocess_cohort_data(cohort_path):
    """
    Load and preprocess the cohort data.
    
    Parameters:
    cohort_path (str): The path to the cohort data CSV file.
    
    Returns:
    pd.DataFrame: The preprocessed cohort data.
    """
    cohort_df = pd.read_csv(cohort_path)
    cohort_df['cohort_month'] = pd.to_datetime(cohort_df['cohort_month'])
    cohort_df['cohort_month'] = cohort_df['cohort_month'].dt.to_period('M')
    cohort_df['month_offset'] = cohort_df['month_offset'].astype(int)
    cohort_filtered = cohort_df[cohort_df['cohort_month'] >= '2017-12']
    cohort_pivot = cohort_filtered.pivot_table(index='cohort_month', columns='month_offset', values='active_user_ratio')
    
    return cohort_pivot

def plot_cohort_heatmap(cohort_pivot):
    """
    Plot a heatmap for the cohort data showing user activity over time.
    
    Parameters:
    cohort_pivot (pd.DataFrame): The pivoted cohort data.
    """
    # Set up the matplotlib figure
    plt.figure(figsize=(12, 8))
    
    # Create the heatmap
    ax = sns.heatmap(cohort_pivot, annot=True, fmt=".0%", cmap='Blues', cbar=False)
    
    # Customize the plot
    plt.title('Monthly Cohorts: User Activity Over Time\n')
    plt.xlabel('Months Since Cohort Start')
    plt.ylabel('Cohort Month')
    plt.yticks(rotation=30)
    
    ax.xaxis.tick_top()
    ax.xaxis.set_label_position('top')
    
    ax.set_yticklabels(ax.get_yticklabels(), rotation=0)
    ax.set_yticklabels([tick.get_text()[:7] for tick in ax.get_yticklabels()])
    
    # Display the heatmap
    plt.show()

# Main function to execute the workflow
def main(cohort_path):
    """
    Main function to execute the workflow.
    
    Parameters:
    cohort_path (str): The path to the cohort data CSV file.
    """
    cohort_pivot = load_and_preprocess_cohort_data(cohort_path)
    plot_cohort_heatmap(cohort_pivot)

# PLot chart
main(cohort_path)
```

## Survival and Segments 

```python
def load_and_preprocess_users_data(users_path):
    """
    Load and preprocess the users data.
    
    Parameters:
    users_path (str): The path to the users data CSV file.
    
    Returns:
    pd.DataFrame: The preprocessed users data.
    """
    users = pd.read_csv(users_path)
    
    # Convert date columns to datetime
    users['activation_date_datatype'] = pd.to_datetime(users['activation_dte'])
    users['last_active_date_datatype'] = pd.to_datetime(users['last_active_date'])
    
    # Calculate duration
    users['duration'] = (users['last_active_date_datatype'] - users['activation_date_datatype']).dt.days
    
    # Create age groups
    bins = [18, 30, 50, 999]
    labels = ['young', 'middle-aged', 'old']
    users['age_group'] = pd.cut(users['age'], bins=bins, labels=labels, right=False)
    
    # Filter users based on activation date
    users_filtered = users[users['activation_date_datatype'] > '2017-12-07']
    
    # Create a 'cohort_month' column
    users_filtered['cohort_month'] = users_filtered['activation_date_datatype'].dt.to_period('M')
    
    return users_filtered

def plot_survival_analysis_by_cohorts(users_filtered):
    """
    Plot survival curves for each cohort.
    
    Parameters:
    users_filtered (pd.DataFrame): The preprocessed users data.
    """
    # Initialize the Kaplan-Meier fitter
    kmf = KaplanMeierFitter()
    
    # Plot survival curves for each cohort
    plt.figure(figsize=(10, 6))
    cohorts = users_filtered['cohort_month'].unique()
    
    for cohort in cohorts:
        cohort_data = users_filtered[users_filtered['cohort_month'] == cohort]
        kmf.fit(durations=cohort_data['duration'], event_observed=cohort_data['churned'], label=str(cohort))
        kmf.plot_survival_function(ci_show=False)
        
    plt.title('Survival Analysis by Monthly Cohorts')
    plt.xlabel('Duration (days)')
    plt.ylabel('Survival Probability')
    plt.legend(title='Cohort Month')
    plt.grid(True)
    plt.show()

# Main function to execute the workflow
def main(users_path):
    """
    Main function to execute the workflow.
    
    Parameters:
    users_path (str): The path to the users data CSV file.
    """
    users_filtered = load_and_preprocess_users_data(users_path)
    plot_survival_analysis_by_cohorts(users_filtered)

# Plot cjart
main(users_path)
```

## Survival Curve per Segment

```python
def plot_survival_analysis(users_filtered, columns_to_plot):
    """
    Plot survival curves for each specified column and perform Log-Rank tests.
    
    Parameters:
    users_filtered (pd.DataFrame): The preprocessed users data.
    columns_to_plot (list): List of columns to plot survival curves for.
    """
    # Create the Kaplan-Meier fitter instance
    kmf = KaplanMeierFitter()
    
    # Plot survival curves for each column
    for column in columns_to_plot:
        plt.figure(figsize=(6, 4))
        
        if column == 'age_group':
            # Handle the 'age_group' column separately
            age_groups = users_filtered[column].unique()
            for age_group in age_groups:
                mask = users_filtered[column] == age_group
                kmf.fit(users_filtered[mask]['duration'], event_observed=users_filtered[mask]['churned'], label=f'Age Group: {age_group}')
                kmf.plot_survival_function(ci_show=False)
            plt.title(f'Survival Analysis by {column}')
        else:
            # Plot for users with value 1
            mask_1 = users_filtered[column] == 1
            kmf.fit(users_filtered[mask_1]['duration'], event_observed=users_filtered[mask_1]['churned'], label=f'{column} = 1')
            kmf.plot_survival_function(ci_show=False)
            
            # Plot for users with value 0
            mask_0 = users_filtered[column] == 0
            kmf.fit(users_filtered[mask_0]['duration'], event_observed=users_filtered[mask_0]['churned'], label=f'{column} = 0')
            kmf.plot_survival_function(ci_show=False)
            
            # Perform Log-Rank test
            results = logrank_test(users_filtered[mask_1]['duration'], users_filtered[mask_0]['duration'], 
                                   event_observed_A=users_filtered[mask_1]['churned'], event_observed_B=users_filtered[mask_0]['churned'])
            stat_sig_text = "There's a statistically significant difference between survival curves of these groups." if results.p_value < .05 else "There's no statistically significant difference between survival curves of these groups."
            plt.title(f'Survival Analysis for {column}\nLog-Rank Test p-value: {results.p_value:.4f}\n{stat_sig_text}')
        
        plt.xlabel('Duration (days)')
        plt.ylabel('Survival Probability')
        plt.legend(title=column)
        plt.grid(True)
        plt.show()

# Main function to execute the workflow
def main(users_filtered, columns_to_plot):
    """
    Main function to execute the workflow.
    
    Parameters:
    users_filtered (pd.DataFrame): The preprocessed users data.
    columns_to_plot (list): List of columns to plot survival curves for.
    """
    users_filtered = load_and_preprocess_users_data(users_path)
    plot_survival_analysis(users_filtered, columns_to_plot)

# Example usage
main(users_filtered, columns_for_survival)
```

## Significance & P-values per Segment

```python
def calculate_churn_ratios(df, column):
    """
    Calculate churn ratios for a given column.
    
    Parameters:
    df (pd.DataFrame): The data frame containing user data.
    column (str): The column for which to calculate churn ratios.
    
    Returns:
    tuple or pd.Series: Churn ratios for the specified column.
    """
    if column == 'age_group':
        churn_rates = df.groupby(column)['churned'].mean()
        return churn_rates
    else:
        churn_rate_1 = df[df[column] == 1]['churned'].mean()
        churn_rate_0 = df[df[column] == 0]['churned'].mean()
        return churn_rate_1, churn_rate_0

def test_significance(df, column):
    """
    Perform a statistical significance test for a given column.
    
    Parameters:
    df (pd.DataFrame): The data frame containing user data.
    column (str): The column for which to perform the significance test.
    
    Returns:
    tuple: Test statistic and p-value for the specified column.
    """
    if column == 'age_group':
        groups = [df[df[column] == group]['churned'] for group in df[column].unique()]
        f_stat, p_value = stats.f_oneway(*groups)
        return f_stat, p_value
    else:
        table = pd.crosstab(df[column], df['churned'])
        chi2, p_value, _, _ = stats.chi2_contingency(table)
        return chi2, p_value

def analyze_churn_ratios(users_filtered, columns_to_analyze):
    """
    Analyze churn ratios and perform statistical significance tests for specified columns.
    
    Parameters:
    users_filtered (pd.DataFrame): The preprocessed users data.
    columns_to_analyze (list): List of columns to analyze.
    """
    for column in columns_to_analyze:
        print(f'{column}:')
        if column == 'age_group':
            churn_rates = calculate_churn_ratios(users_filtered, column)
            f_stat, p_value = test_significance(users_filtered, column)
            
            for age_group, churn_rate in churn_rates.items():
                age_group_name = 'Young' if age_group == 'young' else 'Middle-aged' if age_group == 'middle-aged' else 'Old'
                print(f'  Churn Rate ({age_group_name}): {churn_rate:.2f}')
            print(f'  F-Statistic: {f_stat:.2f}')
        else:
            churn_rate_1, churn_rate_0 = calculate_churn_ratios(users_filtered, column)
            chi2, p_value = test_significance(users_filtered, column)
            
            print(f'  Churn Rate (Value = 1): {churn_rate_1:.2f}')
            print(f'  Churn Rate (Value = 0): {churn_rate_0:.2f}')
            print(f'  Chi-Square Statistic: {chi2:.2f}')
        
        print(f'  p-value: {p_value:.4f}')
        if p_value < 0.05:
            print("  There is a statistically significant difference in churn rates.\n")
        else:
            print("  There is no statistically significant difference in churn rates.\n")

# Main function to execute the workflow
def main(users_filtered, columns_to_analyse):
    """
    Main function to execute the workflow.
    
    Parameters:
    users_filtered (pd.DataFrame): The preprocessed users data.
    columns_to_analyze (list): List of columns to analyze.
    """
    users_filtered = load_and_preprocess_users_data(users_path)
    analyze_churn_ratios(users_filtered, columns_to_analyse)

# Get pvalues and significance
main(users_filtered, columns_for_survival)
```
